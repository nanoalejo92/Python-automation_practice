```python
import pandas as pd
from sqlalchemy import create_engine
from tabulate import tabulate
from credentials import username, password

# connection details
server = "jarbdatabase.database.windows.net"
database = "Azure_EmployeeDB"
username = username
password = password
driver = "ODBC Driver 18 for SQL Server"


# IMPORTANT: use the proper ODBC driver installed on your machine
#driver = "ODBC Driver 18 for SQL Server"

# create SQLAlchemy engine string with Active Directory Interactive auth
conn_str = (
    f"DRIVER={{{driver}}};" 
    f"SERVER={server};"
    f"DATABASE={database};"
    # ðŸ’¡ CHANGE AUTHENTICATION METHOD ðŸ’¡
    "Authentication=ActiveDirectoryPassword;" 
    f"UID={username};"  # Pass the username from credentials.py
    f"PWD={password};"  # Pass the password from credentials.py
    "Encrypt=yes;"
    "TrustServerCertificate=no;"
    # LoginTimeout can be shorter now, e.g., 30 or 60
    "LoginTimeout=60;"
)

# create engine
engine = create_engine(
    f"mssql+pyodbc://{username}:{password}@{server}:1433/{database}?driver={driver}", 
    connect_args={
        "cnxn_str": conn_str, 
        "unicode_results": True,
    }
)

print(f"Connection parameters are set for {database}. Attempting non-interactive login...")

# ... (rest of your try/except block)

# now this is fully supported by pandas
query = "SELECT TOP 10 * FROM [SalesLT].[Customer]"
df = pd.read_sql(query, engine)
print(tabulate(df, headers = 'keys', tablefmt = 'psql'))
```
