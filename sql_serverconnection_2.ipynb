```python
import pyodbc
import pandas as pd
from azure.identity import AzureCliCredential # Or other credential types like DefaultAzureCredential
import struct

server = 'jarbdatabase.database.windows.net'
database = 'Azure_EmployeeDB'

credential = AzureCliCredential() # Authenticates using Azure CLI logged-in user
# For other authentication methods, explore DefaultAzureCredential or other specific credentials
databaseToken = credential.get_token('https://database.windows.net/.default')

tokenb = bytes(databaseToken.token, "UTF-8")
exptoken = b''
for i in tokenb:
    exptoken += bytes({i})
    exptoken += bytes(1)
tokenstruct = struct.pack("=i", len(exptoken)) + exptoken

conn_str = f"Driver={{ODBC Driver 17 for SQL Server}};SERVER={server};DATABASE={database};"
SQL_COPT_SS_ACCESS_TOKEN = 1256 # Constant for access token
conn = pyodbc.connect(conn_str, attrs_before={SQL_COPT_SS_ACCESS_TOKEN: tokenstruct})

query = "SELECT * FROM SalesLT.Customer;"
df = pd.read_sql(query, conn)

conn.close()
```
